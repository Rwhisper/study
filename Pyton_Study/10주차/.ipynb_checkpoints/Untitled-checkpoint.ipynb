```python
input_data = wine[['alcohol', 'sugar', 'pH']]
print(type(input_data))
input_data
```

```python
data = input_data.to_numpy()
print(type(data))
print(data.shape)
data
```

```python
target = wine[['class']].to_numpy()
print(type(target))
print(target.shape)
target
```

```python
target = wine['class'].to_numpy()
print(type(target))
print(target.shape)
target
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_target, test_target = train_test_split(data, target, test_size = 0.2, random_state=42)

print(train_data.shape, test_data.shape)
```

```python
wine.describe()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit(train_data)# 학습 데이터 셋의 평균과 표준편차를 계산

train_scaled = ss.transform(train_data)
test_scaled = ss.transform(train_data)
print(train_scaled)
print(test_scaled)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

kn.fit(train_scaled, train_target)
print('학습 데이터셋 : ', kn.score(train_scaled, train_target))
print('시험 데이터셋 : ', kn.score(test_scaled, test_target))
```
